Binary classification on people who use default payment.

## Data Loading

```python
# Load data from pycaret
from pycaret.datasets import get_data

dataset = get_data('credit')
```

```python
#check the shape of data
dataset.shape
```

```python
# Data for training and validation
data = dataset.sample(frac=0.95, random_state=786)

# Data to be predicted
data_unseen = dataset.drop(data.index)

data.reset_index(inplace=True, drop=True)
data_unseen.reset_index(inplace=True, drop=True)
print('Data for Modeling: ' + str(data.shape))
print('Unseen Data For Predictions: ' + str(data_unseen.shape))
```

## Setup Pycaret Environment

```python
# Setup pycaret environment
from pycaret.classification import *

exp_clf101 = setup(data = data, target = 'default', session_id=123)
```

## Model Evaluation

```python
# Model evaluation
best_model = compare_models()
```

## Create a model

Choice of model and how to access it

```python
models()
```

### Decision Tree

```python
dt = create_model('dt')
```

```python
#trained model object is stored in the variable 'dt'. 
print(dt)
```

### K-nearest-neighbor

```python
knn = create_model('knn')
```

```python
#trained model object is stored in the variable 'dt'. 
print(knn)
```

### Random Forest

```python
rf = create_model('rf')
```

```python
print(rf)
```

## Tune Model

```python
# Auto-tune model using random grid
tuned_dt = tune_model(dt)
print(tuned_dt)

tuned_rf = tune_model(rf)
print(tuned_rf)
```

```python
# Auto tune with custom grid
import numpy as np
tuned_knn = tune_model(knn, custom_grid = {'n_neighbors' : np.arange(0,50,1)})
print(tuned_knn)
```

```python
plot_model(tuned_rf, plot = 'auc')
```

```python
plot_model(tuned_rf, plot = 'pr')
```

```python
plot_model(tuned_rf, plot='feature')
```

```python
plot_model(tuned_rf, plot='feature')
```

```python
evaluate_model(tuned_rf)
```

## Pedict model on test data that has automatically seperated from train data set

```python
predict_model(tuned_rf)
```

### Finalize the model

```python
# Fits the model to the remaining test data set
final_rf = finalize_model(tuned_rf)

# Final Random Forest model parameters for deployment
print(final_rf)

# Re-predict the data
predict_model(final_rf)
```

## Predict unseen data

```python
unseen_predictions = predict_model(final_rf, data=data_unseen)
unseen_predictions.head()
```

```python
from pycaret.utils import check_metric
check_metric(unseen_predictions['default'], unseen_predictions['Label'], metric = 'Accuracy')
```

## Saving and Loading model

```python
# Save Model
save_model(final_rf,'rf_model')
```

```python
# Load Model
saved_final_rf = load_model('rf_model')
```

```python
new_prediction = predict_model(saved_final_rf, data=data_unseen)
new_prediction.head()

from pycaret.utils import check_metric
check_metric(new_prediction['default'], new_prediction['Label'], metric = 'Accuracy')
```
